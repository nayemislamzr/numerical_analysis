```python
import numpy as np
import pandas as pd

def f(x):
    return x**3-(0.165*x**2)+3.993*1e-4


def bisection(l, h, tol):
    root = None
    stk = []
    itr = 1

    xl, xu, xm, error, f_xm = ([] for i in range(5))

    while True:
        m = (l+h)/2

        xl.append(l)
        xu.append(h)
        xm.append(m)
        f_xm.append(f(m))

        if root != None:
            ap_error = np.abs((m-root)/m)*100
            error.append(ap_error)
            if ap_error <= tol:
                break
        else:
            error.append("NAN")

        val = f(l)*f(m)
        root = m

        if val < 0:
            h = m
        elif val > 0:
            l = m
        else:
            break

    d = {"xl" : xl, "xu" : xu, "xm" : xm, "error(%)" : error, "f(m)" : f_xm}
    tabledf = pd.DataFrame(data = d)
    display(tabledf)

def main():
    bisection(0.00, 0.11, 0.0001)


if __name__ == "__main__":
    main()
```
